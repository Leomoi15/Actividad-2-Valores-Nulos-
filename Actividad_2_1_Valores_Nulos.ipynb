**Métodos de Sustitución de Valores Nulos**

```python
#Realizamos una copia del dataframe para poder tener un backup en caso de algun error y no tener que volver a cargar los datos
data2=data.copy()
```

```python
#Primer método de sustitución de valores nulos
#Sustituir valores nulos con promedio o media, donde para este ejemplo seleccionamos la columna de salon_ventas
data2["salon_ventas"]=data2["salon_ventas"].fillna(round(data2["salon_ventas"].mean(),1))
```

```python
#Corroboramos valores nulos para ver si la columna seleccionada ya esta corregida
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
Eliminacion2 = data2.to_csv("Entregable2.csv")
```

```python
from google.colab import files

files.download("Entregable2.csv")
```

```python
#Segundo método de sustitución de valores nulos
#Sustituir valores nulos con mediana, donde para este ejemplo seleccionamos la columna de tarjetas_debito
data2["tarjetas_debito"]=data2["tarjetas_debito"].fillna(round(data2["tarjetas_debito"].median(),1))
data2
```

```python
#Corroboramos valores nulos para ver si la columna seleccionada ya esta corregida
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
Eliminacion3 = data2.to_csv("Entregable3.csv")
```

```python
#Tercer método de sustitución de valores nulos
#Sustituir valores nulos por un valor numérico en  concreto, donde seleccione la columna otros_medios, y seleccione un dato al azar el cual fue utilizado para sustituir los valores nulos
data2["otros_medios"]= data2["otros_medios"].fillna(1909221.374)
data2
```

```python
#Cuarto método de sustitución de valores nulos
#Sustituir valores nulos por un string en  concreto, donde aqui no es necesario ya que las fechas estan completas, pero asi quedaria para nuestro dataframe
#data2["indice_tiempo"] =data3["indice_tiempo"].fillna("2030-06-19 00:00:00")
```

```python
#Quinto método de sustitución de valores nulos
#Sustituir valores nulos para llenar los valores nulos en la columna "bebidas" con el valor no nulo más cercano hacia adelante en la misma columna.
#Filtro por columnas
data2["bebidas"] =data2["bebidas"].fillna(method="ffill")
```

**Método de Eliminación de valores nulos**

```python
#Identificar valores nulos por columna de nuestro dataframe original
valores_nulos=data.isnull().sum()
valores_nulos
```

```python
#Realizamos una copia del dataframe para poder trabajar de mejor manera
data3=data.copy()
```

```python
#Eliminamos filas con nulos de la copia de nuestro dataframe
#0 para eliminar filar y 1 para eliminar columnas, donde creamos un nuevo dataframe para guardar los nuevos valores
data4= data3.dropna(0)
data4
```

```python
#Identificar valores nulos por columna
valores_nulos=data4.isnull().sum()
valores_nulos
```
